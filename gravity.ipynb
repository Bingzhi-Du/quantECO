```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_patent = pd.read_excel('/Users/bingzhi/git/eco/quantECO/patent.xlsx')
```

```python
df_gravity = pd.read_csv('/Users/bingzhi/git/eco/quantECO/Gravity_csv_V202211/Gravity_V202211.csv')
```

```python
df_patent.columns = ['year', 'KOR', 'DEU', 'USA', 'JPN', 'CHN']
df_patent
```

```python
filtered_df_gravity = df_gravity[(df_gravity['year'] >= 2005) & (df_gravity['year'] <= 2020)]
```

```python
filtered_df_gravity_2 = filtered_df_gravity[filtered_df_gravity['country_id_o'].isin(['KOR', 'DEU', 'USA', 'JPN', 'CHN'])]
```

```python
columns = ['year', 'country_id_o', 'country_id_d', 'pop_o', 'pop_d', 'gdp_o', 'gdp_d', 'gdpcap_o', 'gdpcap_d', 'wto_o', 'wto_d', 'gatt_o', 'gatt_d', 'fta_wto_raw', 'distcap','tradeflow_imf_o']
filtered_df_gravity_3 = filtered_df_gravity_2[columns]
filtered_df_gravity_3
```

```python
filtered_df_gravity_3.to_excel('filtered_gravity_data.xlsx', index=False)

```

```python
filtered_df_gravity_3
```

```python
countries_with_na = filtered_df_gravity_3[
    filtered_df_gravity_3[['pop_d', 'gdp_d', 'gdpcap_d','tradeflow_imf_o']].isna().any(axis=1)
]['country_id_d'].unique()
```

```python
countries_with_na
```

```python
filtered_df_gravity_4 = filtered_df_gravity_3[~filtered_df_gravity_3['country_id_d'].isin(countries_with_na)]
```

```python
filtered_df_gravity_4
```

```python
df_patent_long = pd.melt(df_patent, id_vars=['year'], var_name='country_id_o', value_name='patent_o')

# 合并数据框
filtered_df_gravity_5 = pd.merge(filtered_df_gravity_4, df_patent_long, on=['year', 'country_id_o'], how='left')
```
